### Column names fix: ###

```python
# Regex substituion with positive lookahead and lookbehind that puts underscore at every place that:
# is not not the start/preceeded by only the start symbol (?<!^)
# and is followed by a capital letter (?=[A-Z]) 
# that is NOT preceeded by another capital letter (?<![A-Z] ..capital letter.. )
add_underscores = lambda string: re.sub(r'(?<!^)(?=(?<![A-Z])[A-Z])', '_', string)

new_col_names = [add_underscores(col).lower() for col in asthma.columns]
asthma.columns = new_col_names
asthma.dtypes
```

### Column Categories: ###
In the helpers.py I've made this function return a new df and not do modifications 'inplace'

```python
def set_cols_dtype(df, cols, dtype):
    for col in cols:
        if dtype == 'numeric':
            df[col] = pd.to_numeric(df[col], errors = 'coerce')
        else:
            df[col] = df[col].astype(dtype)
```

### Fature Manipulation: ###

```python
asthma.dtypes
```

```python
def normalize(col):
    start_at_zero = col - col.min()
    return start_at_zero / start_at_zero.max()
```

In the helpers.py I've made this function return a new df and not do modifications 'inplace'

```python
def encode_cols(df, cols):
    cat_encoder = LabelEncoder()
    for col in cols:
        df[col] = cat_encoder.fit_transform(df[col])
```

```python
# Useless step because they are already numbers...
encode_cols(asthma, bool_cols)
encode_cols(asthma, category_cols)
```

```python
for col in numeric_cols.columns:
    asthma[col] = normalize(asthma[col])
```

#### Categorical features: ####

```python
cat_bool_cols = category_cols + bool_cols
chi2_cols = asthma[cat_bool_cols].drop(columns = ['diagnosis'])
chi2_stats, p_values = chi2(chi2_cols, asthma.diagnosis)
```

Small deviations and big p-values, means that it is not likely that any of these features is a strong indicator of target.

```python
chi2_stats, p_values
```

This is the feature with the highest deviation from the expected and lowest p-value, the highest chance that it might have an influence on target. 
They would no be considered significant normally but it's the best Ican work with for now. 

```python
ind = chi2_stats.argmax()
print (cat_bool_cols[ind])
print ('chi2: ', chi2_stats[ind], '\np-value: ', p_values[ind])
```

#### Continuous features: ####
